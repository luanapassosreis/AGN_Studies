## Leptons

```python
rate_ebremss = rate_bremss_e(n_c, 1, Ee)
rate_IC_X = rate_IC(Ee, eps_x, nph_x)
rate_IC_OUV = rate_IC(Ee, eps_OUV, nph_OUV)
```

```python
## replace zeros with NaN

rate_ebremss = np.where(rate_ebremss > 0, rate_ebremss, np.nan)
rate_IC_X = np.where(rate_IC_X > 0, rate_IC_X, np.nan)
rate_IC_OUV = np.where(rate_IC_OUV > 0, rate_IC_OUV, np.nan)
```

```python
t_loss_esynch = 1 / rate_synch(Ee, B_c, me)
t_loss_ebrems = 1 / rate_ebremss
t_loss_IC_X = 1 / rate_IC_X
t_loss_IC_OUV = 1 / rate_IC_OUV
```

```python
t_loss_ebrems = np.nan_to_num(t_loss_ebrems, nan=np.inf)
t_loss_IC_X = np.nan_to_num(t_loss_IC_X, nan=np.inf)
t_loss_IC_OUV = np.nan_to_num(t_loss_IC_OUV, nan=np.inf)
```

```python
t_eloss_total = t_loss_esynch + t_loss_ebrems + t_loss_IC_X + t_loss_IC_OUV
```

```python
## leptons

plt.figure(figsize=(10, 6))

## acceleration
plt.loglog(Ee/eV,[(me/mp)*t_acc_fermi(va, delta_RX)]*len(Ee),'--', label='Fermi Acceleration', color='black')
plt.loglog(Ee_drift/eV,t_acc_drift(Ee_drift, B_c, va),'--', label='Drift Acceleration', color='gray')

## losses
plt.loglog(Ee/eV,t_loss_ebrems,label='e-Bremsstrahlung',color='purple')
plt.loglog(Ee/eV,t_loss_esynch,label='e-Synchrotron',color='blue')
plt.loglog(Ee/eV,t_loss_IC_X,label='IC (X)',color='magenta')
plt.loglog(Ee/eV,t_loss_IC_OUV,label='IC (OUV)',color='cyan')

plt.loglog(Ee/eV, t_eloss_total, label='Total Loss Timescale', linestyle='-.', color='red')

# plt.xlim((1e9,Ep.max()/eV))
# plt.ylim((1e-11,1e6))

plt.xlabel('Energy [eV]')
plt.ylabel('Timescale [s]')
plt.title('Leptons Cooling & Acceleration Timescales')

plt.legend(frameon=False,loc='lower right')
plt.savefig('plots/timescales_leptons_eV.png',dpi=300)

plt.grid(True, which='both', color='gray', linestyle='--', linewidth=0.5, alpha=0.3)
plt.show()
```

```python
rate_psynch = rate_synch(Ep, B_c, mp)

rate_pp = rate_p_p(n_c, Ep)
rate_pgamma_X = rate_p_gamma(Ep, eps_x, nph_x)
rate_pgamma_OUV = rate_p_gamma(Ep, eps_OUV, nph_OUV)
```

```python
## replace zeros with NaN

rate_pp = np.where(rate_pp > 0, rate_pp, np.nan)
rate_pgamma_X = np.where(rate_pgamma_X > 0, rate_pgamma_X, np.nan)
rate_pgamma_OUV = np.where(rate_pgamma_OUV > 0, rate_pgamma_OUV, np.nan)
```

```python
t_loss_psynch = 1 / rate_psynch
t_loss_pp = 1 / rate_pp
t_loss_pgamma_X = 1 / rate_pgamma_X
t_loss_pgamma_OUV = 1 / rate_pgamma_OUV
```

```python
t_loss_pp = np.nan_to_num(t_loss_pp, nan=np.inf)
t_loss_pgamma_X = np.nan_to_num(t_loss_pgamma_X, nan=np.inf)
t_loss_pgamma_OUV = np.nan_to_num(t_loss_pgamma_OUV, nan=np.inf)
```

```python
t_ploss_total = t_loss_psynch + t_loss_pp + t_loss_pgamma_X + t_loss_pgamma_OUV
```

```python
## hadrons

plt.figure(figsize=(10, 6))

## acceleration
plt.loglog(Ep/eV,[t_acc_fermi(va, delta_RX)]*len(Ep),'--', label='Fermi Acceleration',color='black')
plt.loglog(Ep_drift/eV,t_acc_drift(Ep_drift, B_c, va),'--', label='Drift Acceleration',color='gray')

## losses
plt.loglog(Ep/eV,t_loss_psynch,label='p-Synchrotron',color='blue')
plt.loglog(Ep/eV,t_loss_pgamma_X,'--',label='P-gamma (X)',color='orange')
plt.loglog(Ep/eV,t_loss_pgamma_OUV,label='P-gamma (OUV)',color='cyan')

# Apply the condition for plotting only after 10^8 eV
Ep_filtered = Ep[Ep/eV > 10**(8.37)]
t_loss_pp_filtered = t_loss_pp[Ep/eV > 10**(8.37)]
plt.loglog(Ep_filtered/eV, t_loss_pp_filtered, label='P-P', color='magenta')
# plt.loglog(Ep/eV,t_loss_pp,label='P-P',color='magenta')

plt.loglog(Ep/eV, t_ploss_total, label='Total Loss Timescale', linestyle='-.', color='red')

# plt.loglog(Ep/eV,rate_p_gamma(Ep, eps_x, nph_x),'--',label='P-gamma (X)',color='magenta')
# plt.loglog(Ep/eV,rate_p_gamma(Ep, eps_OUV, nph_OUV),label='P-gamma (OUV)',color='cyan')
# # plt.loglog(Ep/eV,rate_pg_cool(Ep, eps_x, nph_x),label='P-gamma (X)', color='red')
# # plt.loglog(Ep/eV,rate_pg_cool(Ep, eps_OUV, nph_OUV),label='P-gamma (OUV)',color='blue')

# plt.loglog(Ep/eV,rate_bth_cool(Ep, eps_x, nph_x),label='B-H (X)', color='magenta')
# plt.loglog(Ep/eV,rate_bth_cool(Ep, eps_OUV, nph_OUV),'--',label='B-H (OUV)', color='cyan')

# plt.loglog(Ep/eV,cooling_rate_bth_isotropic,label='B-H (X)',color='orange')
# plt.loglog(Ep/eV,cooling_rate_bth_ou,label='B-H (OUV)',color='purple')

# plt.xlim((1e7,1e20))
# plt.ylim((1e-14,1e11))

plt.xlabel('Energy [eV]')
plt.ylabel('Timescale [s]')
plt.title('Hadrons Cooling & Acceleration Timescales')

plt.legend(frameon=False,loc='upper right')
# plt.xticks([1e9,1e10,1e11,1e12,1e13,1e14,1e15],labels=['9','10','11','12','13','14','15'])
plt.savefig('plots/timescales_hadrons_eV.png',dpi=300)

plt.grid(True, which='both', color='gray', linestyle='--', linewidth=0.5, alpha=0.3)
plt.show()
```

```python
## testing the function for IC

def test_IC(eps):
    gamma = 4 * eps * Ee / mec2**2
    return (gamma * Ee) / (1 + gamma)

```

```python
upper_X = test_IC(eps_x)
upper_OUV = test_IC(eps_OUV)
```

```python
# Plot results
plt.figure(figsize=(8, 6))

plt.loglog(eps_x, upper_X, label=r'$\epsilon_X$', color='blue')
plt.loglog(eps_OUV, upper_OUV, label=r'$\epsilon_{OUV}$', color='red')

plt.xlabel(r'Photon Energy $\epsilon$ [eV]')
plt.ylabel(r'Test IC Function Value')
plt.title('Testing the limits for IC Values for $\epsilon_X$ and $\epsilon_{OUV}$')
plt.legend()

plt.grid(True, which='both', color='gray', linestyle='--', linewidth=0.5, alpha=0.3)
plt.show()
```

```python
# Compute the function for both cases
test_x = np.array([test_IC(eps) for eps in eps_x])  # X-ray range
test_OUV = np.array([test_IC(eps) for eps in eps_OUV])  # Optical-UV range

# Plot results
plt.figure(figsize=(8, 6))

for i in range(len(eps_x)):
    plt.loglog(Ee, test_x[i], 'b-', alpha=0.1)  # X-ray range
for i in range(len(eps_OUV)):
    plt.loglog(Ee, test_OUV[i], 'r-', alpha=0.1)  # Optical-UV range

plt.xlabel(r'$E_e$ [eV]')
plt.ylabel(r'Test IC function')
plt.title('Testing IC Limits for $\epsilon_X$ and $\epsilon_{OUV}$')
plt.legend(['X-ray', 'Optical-UV'])
plt.grid(True, which='both', linestyle='--', alpha=0.5)

plt.show()
```
